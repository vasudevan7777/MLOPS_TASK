### 1. Create YOLO dataset folder structure

```python
import os

dataset_path = '/content/dataset'

# Create base directories
os.makedirs(f'{dataset_path}/images/train', exist_ok=True)
os.makedirs(f'{dataset_path}/images/val', exist_ok=True)
os.makedirs(f'{dataset_path}/labels/train', exist_ok=True)
os.makedirs(f'{dataset_path}/labels/val', exist_ok=True)

print("Dataset folder structure created successfully.")
```

### 2. Split dataset into 80% train and 20% validation automatically
### 3. Copy images and txt labels into correct train and val folders

```python
!pip install scikit-learn
```

```python
import shutil
from sklearn.model_selection import train_test_split

# Define source directories
image_source_dir = '/content/drive/MyDrive/SMART_RAILWAY_YOLO_PROJECT/batch1_images'
label_source_dir = '/content/runs/detect/predict/labels'

# Get all image files and their base names
image_files = [f for f in os.listdir(image_source_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
image_basenames = {os.path.splitext(f)[0]: f for f in image_files}

# Get all label files and their base names
label_files = [f for f in os.listdir(label_source_dir) if f.lower().endswith('.txt')]
label_basenames = {os.path.splitext(f)[0]: f for f in label_files}

# Pair images with labels that have matching basenames
dataset_items = []
for basename, image_filename in image_basenames.items():
    if basename in label_basenames:
        dataset_items.append({
            'image_path': os.path.join(image_source_dir, image_filename),
            'label_path': os.path.join(label_source_dir, label_basenames[basename])
        })

print(f"Found {len(dataset_items)} paired image-label sets.")

if not dataset_items:
    print("No matching image and label files found. Please check your source directories.")
else:
    # Split the dataset into training (80%) and validation (20%)
    train_data, val_data = train_test_split(dataset_items, test_size=0.2, random_state=42)

    print(f"Training set size: {len(train_data)}")
    print(f"Validation set size: {len(val_data)}")

    # Copy files to their respective directories
    def copy_files(data_list, image_dest_dir, label_dest_dir):
        for item in data_list:
            shutil.copy(item['image_path'], image_dest_dir)
            shutil.copy(item['label_path'], label_dest_dir)

    # Destination directories
    train_images_dir = f'{dataset_path}/images/train'
    train_labels_dir = f'{dataset_path}/labels/train'
    val_images_dir = f'{dataset_path}/images/val'
    val_labels_dir = f'{dataset_path}/labels/val'

    print("Copying training files...")
    copy_files(train_data, train_images_dir, train_labels_dir)
    print("Copying validation files...")
    copy_files(val_data, val_images_dir, val_labels_dir)

    print("Dataset split and copying complete.")
```

### 4. Generate correct data.yaml file for YOLOv8

```python
import yaml

# Define the path for the data.yaml file
data_yaml_path = f'{dataset_path}/data.yaml'

# Define the content of the data.yaml file
data_content = {
    'path': dataset_path, # Root path to the dataset
    'train': 'images/train',  # Path to training images relative to 'path'
    'val': 'images/val',    # Path to validation images relative to 'path'
    'nc': 1,  # Number of classes
    'names': ['person']  # Class names
}

# Write the content to the data.yaml file
with open(data_yaml_path, 'w') as f:
    yaml.dump(data_content, f, default_flow_style=False)

print(f"data.yaml created at: {data_yaml_path}")

# Display the content of the data.yaml file for verification
with open(data_yaml_path, 'r') as f:
    print("\nContent of data.yaml:")
    print(f.read())
```

### 5. Load YOLOv8 model

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8n model
model = YOLO('yolov8n.pt')

print("YOLOv8 model loaded successfully.")
```

### 6. Train the YOLO model

Now we'll train the YOLOv8 model using the dataset we prepared and the fast settings you requested. This will create a lightweight model quickly.

```python
results = model.train(data=data_yaml_path, epochs=10, imgsz=320, batch=8, workers=2)

print("Model training complete.")
```

### 7. Test prediction on one image after training

Let's use the newly trained model (`best.pt`) to make a prediction on a sample image from your `test_images` folder to see how it performs.

```python
import glob
import os
from ultralytics import YOLO

# Get a list of all image files in the test_images folder
test_images_path = '/content/drive/MyDrive/SMART_RAILWAY_YOLO_PROJECT/test_images'
sample_images = glob.glob(os.path.join(test_images_path, '*.jpg')) # Adjust if you have other image types

if sample_images:
    sample_image_path = sample_images[0] # Take the first image as a sample
    print(f"Predicting on sample image: {sample_image_path}")

    # The 'results' object from model.train() contains the save directory
    # We need to ensure 'results' is available. If the notebook was restarted,
    # 'results' might not be in scope. For robustness, we can infer the path.
    # Assuming the default save_dir is 'runs/detect/train'
    # If you run model.train() multiple times, it will create 'train2', 'train3', etc.
    # For this demonstration, we'll assume the most recent 'train' directory.

    # Get the latest run directory if 'results' is not directly available
    try:
        best_model_path = os.path.join(results.save_dir, 'weights', 'best.pt')
    except NameError:
        # If 'results' object is not defined (e.g., after a kernel restart)
        # We need to find the latest training run directory programmatically
        runs_dir = '/content/runs/detect'
        train_dirs = [d for d in os.listdir(runs_dir) if d.startswith('train') and os.path.isdir(os.path.join(runs_dir, d))]
        train_dirs.sort(key=lambda x: int(x[len('train'):] or 0), reverse=True) # Sort by number, 'train' before 'train1'
        if train_dirs:
            latest_train_dir = os.path.join(runs_dir, train_dirs[0])
            best_model_path = os.path.join(latest_train_dir, 'weights', 'best.pt')
            print(f"Inferred best.pt path: {best_model_path}")
        else:
            print("Could not find any training run directories to locate best.pt. Please ensure training was successful.")
            best_model_path = None

    if best_model_path and os.path.exists(best_model_path):
        # Load the best trained model
        trained_model = YOLO(best_model_path)

        # Perform prediction
        trained_model.predict(
            source=sample_image_path,
            save=True,
            save_txt=True,
            classes=[0] # Assuming 'person' is class 0 based on your data.yaml
        )
    else:
        print("Best model (best.pt) not found. Please ensure the training step was executed successfully.")
else:
    print("No sample images found in the test_images folder for prediction. Please ensure test_images folder exists and contains images.")
```

### 8. Show where the `best.pt` model is saved

The trained model weights (`best.pt` and `last.pt`) are saved in the `runs` directory created by Ultralytics. The `best.pt` file represents the best performing model during training.

```python
import os

# The path to the directory where results were saved during training
# This path is dynamically available from the 'results' object after training
# If 'results' is not available (e.g., after kernel restart), we infer it.

try:
    model_save_dir = results.save_dir
except NameError:
    # If 'results' object is not defined (e.g., after a kernel restart)
    # We need to find the latest training run directory programmatically
    runs_dir = '/content/runs/detect'
    train_dirs = [d for d in os.listdir(runs_dir) if d.startswith('train') and os.path.isdir(os.path.join(runs_dir, d))]
    train_dirs.sort(key=lambda x: int(x[len('train'):] or 0), reverse=True) # Sort by number, 'train' before 'train1'
    if train_dirs:
        model_save_dir = os.path.join(runs_dir, train_dirs[0])
        print(f"Inferred model save directory: {model_save_dir}")
    else:
        model_save_dir = None
        print("Could not find any training run directories.")

if model_save_dir and os.path.exists(model_save_dir):
    best_pt_path = os.path.join(model_save_dir, 'weights', 'best.pt')
    last_pt_path = os.path.join(model_save_dir, 'weights', 'last.pt')

    print(f"Best model (best.pt) saved at: {best_pt_path}")
    print(f"Last model (last.pt) saved at: {last_pt_path}")

    # You can also list the contents of the weights directory
    weights_dir = os.path.join(model_save_dir, 'weights')
    if os.path.exists(weights_dir):
        print(f"\nContents of {weights_dir}:")
        !ls -lh {weights_dir}
    else:
        print(f"Weights directory not found: {weights_dir}")
else:
    print("Model save directory not found. Please ensure the training step was executed successfully.")
```
